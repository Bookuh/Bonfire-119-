### **Task 1: List Comprehension**
Given a list as a parameter, write a function named `less_than_10`, that uses list comprehension to returns the input list with all numbers less than 10.

```python
def less_than_10(a_list):
    new_list = []
    for new in a_list:
        if new < 10:
            new_list.append(new)
    return new_list        
test_list = [12, 9, 18, 8, 4, 5, 27, 1, 13, 2]
less_than_10(test_list)
```

### **Task 2: Merge**
Write a function named `merge_lists` that takes two seperate lists as arguments and returns the two lists merged together AND sorted.

```python
def merge_list(list_1, list_2):
    combined = list_1 + list_2
    combined.sort()
    return combined 
test_list_1 = [2, 23, 54, 84, 98, 12, 43, 23, 45, 21, 2233, 34, 434]
test_list_2 = [25, 42, 24, 434, 475, 91]
print(merge_list(test_list_1, test_list_2))
```

### **Task 3.2: Range Function**
Using the three lists you created in the last task, build a function named `name_generator` which takes in each of these lists as an input and performs the following:

* Iterates through each list using a range
* Creates a `full_name` variable, which has a first, middle, and last name
* Saves the `full_name` to the list `full_names`
* Returns the `full_names` list

**HINT** Your range should not be a hard-coded number

```python
def name_generator(name_1, name_2, name_3):
    full_names = []
    for index in range(len(name_1)):
        full_name = f"{name_1[index].title()}, {name_2[index].title()}, {name_3[index].title()}"
        full_names.append(full_name)
    return full_names

first_names = ['jim', 'john', 'david', 'dwayne', 'chris', 'kevin', 'jordan', 'sean', 'mark', 'devin' ]
middle_names = ['faith', 'jewel', 'marie', 'taylor', 'jacob', 'star', 'autumn', 'destiny', 'ann', 'alexander']
last_names = ['wilbert,', 'andrews', 'jackson', 'johnson', 'romero', 'james', 'murray', 'irving', 'manning', 'harris']
name_generator(first_names, middle_names, last_names)
    
   
```
